```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("mcycle.csv")
```

```python
df
```

```python
from sklearn import linear_model
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Linear Regression
X=df[["times"]]
y=df["accel"]
lin_reg=linear_model.LinearRegression()
lin_reg.fit(X,y)
```

```python
# Decision Tree Regressor
tree_reg=tree.DecisionTreeRegressor()
tree_reg.fit(X,y)
```

```python
time_range=pd.DataFrame(np.linspace(X.min(),X.max(),100),columns=["times"])
time_range
```

```python
plt.xlabel("times")
plt.ylabel("accel")
plt.scatter(X,y,color="black")
plt.plot(time_range,lin_reg.predict(time_range),color="red",label="Lineare Regression")
plt.plot(time_range,tree_reg.predict(time_range),color="green",label=" Regressionbaum")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import sklearn.datasets as skd

X, y = skd.make_moons(
    n_samples = 1000,
    noise = 0.2,
    random_state = 42
)

plt.scatter(*X.T, c=y,s=8,cmap=plt.cm.Spectral)
plt.ylabel(r'$x_2$')
plt.xlabel(r'$x_1$')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
# Logistic Regression
log_reg=LogisticRegression()
log_reg.fit(X,y)
```

```python
# Klassificationbaum
cls_baum=DecisionTreeClassifier()
cls_baum.fit(X,y)
```
